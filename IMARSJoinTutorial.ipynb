```python
import pandas as pd
import geopandas as gpd
import numpy as np
```

```python
#import the four data files (in the folder the files are in excel format, they need to be converted to csv )
commandAddress=pd.read_csv('NPS GOccIvPA Command Address 4-2018 to 6-2021.csv')
personReport=pd.read_csv('NPS MVCOccGPersonReport Command 4-2018 to 6-2021.csv')
reportCommand=pd.read_csv('NPS MVCOccReportCommand_Classification 4-2018 to 6-2021.csv')
vehicleReport=pd.read_csv('NPS MVCOccVehicleReport Command 4-2018 to 6-2021.csv')
```

```python
# change the name of crash date and time to a unique value for each df
#later we'll use this to check the accuracy of the merge 
commandAddress.rename(columns={'Crash_Date_Time':'Crash_Date_Time_address'}, inplace=True)
personReport.rename(columns={'Crash_Date_Time':'Crash_Date_Time_person'}, inplace=True)
reportCommand.rename(columns={'Crash_Date_Time':'Crash_Date_Time_report'}, inplace=True)
vehicleReport.rename(columns={'Crash_Date_Time':'Crash_Date_Time_vehicle'}, inplace=True)
```

```python
# here we use merge on the record number to create a new df of both the person file and address file
addressPerson=commandAddress.merge(personReport, how='left', on='IMARS_Record_No')
addressPerson
```

```python
# now I want to check that the merge was correct. 
#I'm using the crash date column since this is another shared attribute between the files
addressPerson['check1']=np.where((addressPerson['Crash_Date_Time_address']==addressPerson['Crash_Date_Time_person']), 0, 1)
addressPerson['check1'].sum()
```

```python
# this means that 13 rows had different crash times for the same record
#lets look into these more with a df that is only these rows
addressPersonWrongDate=addressPerson[addressPerson['check1']==1]
pd.set_option('display.max_columns', None)
addressPersonWrongDate.insert(2, 'Crash_Date_Time_person', addressPersonWrongDate['Crash_Date_Time_person'], allow_duplicates = True)
addressPersonWrongDate
```

```python
# since the columns are mostly the same, the differences are very small, we'll keep merging the other files 
addressPersonReport=addressPerson.merge(reportCommand, how='left', on='IMARS_Record_No')
```

```python
#checking the date time again
addressPersonReport['check2']=np.where((addressPersonReport['Crash_Date_Time_address']==addressPersonReport['Crash_Date_Time_report']), 0, 1)
addressPersonReport['check2'].sum()
```

```python
#looking at the 29 differences
addressPersonReportWrongDate=addressPersonReport[addressPersonReport['check2']==1]
pd.set_option('display.max_columns', None)
addressPersonReportWrongDate.insert(2, 'Crash_Date_Time_report', addressPersonReportWrongDate['Crash_Date_Time_report'], allow_duplicates = True)
addressPersonReportWrongDate
```

```python
# they still look mostly the same, now we'll do the last merge 
addressPersonReportVehicle=addressPersonReport.merge(vehicleReport, how='left', on='IMARS_Record_No')
```

```python
#checking the date time again
addressPersonReportVehicle['check3']=np.where((addressPersonReportVehicle['Crash_Date_Time_address']==addressPersonReportVehicle['Crash_Date_Time_vehicle']), 0, 1)
addressPersonReportVehicle['check3'].sum()
```

```python
#looking at the 92 differences
addressPersonReportVehicleWrongDate=addressPersonReportVehicle[addressPersonReportVehicle['check3']==1]
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
addressPersonReportVehicleWrongDate.insert(2, 'Crash_Date_Time_vehicle', addressPersonReportVehicleWrongDate['Crash_Date_Time_vehicle'], allow_duplicates = True)
addressPersonReportVehicleWrongDate
```
