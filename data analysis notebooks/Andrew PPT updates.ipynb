```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
from matplotlib.ticker import PercentFormatter
import numpy as np
import math
import seaborn as sns
import os
```

```python
crash_df = pd.read_csv("./crash_data_joined.csv")
```

```python
crash_df.head()
```

```python
crash_df['CRASH_DATE']=pd.to_datetime(crash_df['CRASH_DATE'])
crash_df['day_of_week'] = crash_df['CRASH_DATE'].dt.day_name
crash_df['year'] = crash_df['CRASH_DATE'].dt.year
```

```python
crash_df.year.value_counts()
```

```python
crash_df = crash_df.loc[crash_df['year']>1989]
crash_df = crash_df.loc[crash_df['year']<2021]
```

```python
crash_df_IMR = crash_df.loc[crash_df['REGION']=="IM"]
```

```python
crash_df_IMR.head()
```

```python
crash_df_IMR.REGION.value_counts()
```

```python
crash_df_IMR.Fatality.value_counts()
```

```python
crash_df_IMR_Fatal = crash_df_IMR.loc[crash_df_IMR['FATALS'] > 0.5]
```

```python
crash_df_IMR_Fatal.head()
```

```python
crash_df_IMR_Fatal.Fatality.value_counts()
```

```python
crash_df_IMR_Fatal.columns
```

```python
column_sum = crash_df_IMR_Fatal["Collision with Other Motor Vehicle"].sum()
print(column_sum)
```

```python
Crash_types = ['Non-Collision', 'Collision with Other Motor Vehicle', 'Collision with Fixed Object',
       'Collision with Pedestrian', 'Collision with Bicycle',
       'Collision with Parked Motor Vehicle', 'Collision with Railway Train',
       'Collision with Animal', 'Collision with Other Object',
       'Collision with Unknown', 'Other Accident Class']

second=[]

for type in Crash_types:
    m = crash_df_IMR_Fatal[type].sum()
    second.append(m)
    
print(second)


df_fatal = pd.DataFrame(list(zip(Crash_types,second)),
               columns =['Crash Classification', 'Number of Crashes'])

df_fatal.head(15)
```

```python
df_fatal['% of Crashes'] = round((df_fatal["Number of Crashes"] / df_fatal["Number of Crashes"].sum())*100,1)
df_fatal['% of Crashes'] = df_fatal['% of Crashes'].astype(str)+'%'
```

```python
df_fatal.sort_values(by='Number of Crashes', ascending=False)
```
